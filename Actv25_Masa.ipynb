```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Paso 1

Se utilizó la aplicación web [WebPlotDigitizer](https://automeris.io/WebPlotDigitizer/) para extraer datos de la CDE proporcionada y replicarla en Python.

```python
data = pd.read_csv('Actv25_Data.csv', sep=';', header=None) # Puntos X,Y de la CDE
data.columns = ['X', 'Y']
X = np.array(data.X)    # Puntos X
Y = np.array(data.Y)    # Puntos Y
```

```python
plt.figure(figsize=(10, 10))
plt.plot(X, Y, label='CDE')
plt.title('CDE')
plt.xlabel('X (Fracción molar de Benceno en el líquido')
plt.ylabel('Y (Fracción molar de Benceno en el gas')
plt.legend()
plt.grid(True)
```

# Paso 2

```python
aux_X = [X[0], X[-1]]  # Puntos X para línea 45°   
aux_Y = [Y[0], Y[-1]]  # Puntos Y para línea 45°

plt.figure(figsize=(10, 10))
plt.plot(X, Y, label='CDE')
plt.plot(aux_X, aux_Y, 'r', label='Línea aux 45°')
plt.title('CDE')
plt.xlabel('X (Fracción molar de Benceno en el líquido')
plt.ylabel('Y (Fracción molar de Benceno en el gas')
plt.grid(True)
plt.legend()
```

# Paso 3

```python
zF = 0.6    # Fracción molar de Benceno en el feed
xB = 0.03   # Fracción molar de Benceno en los fondos
xD = 0.96   # Fracción molar de Benceno en el destilado
print(f'La fracción molar de Benceno en el feed es {zF}')
print(f'La fracción molar de Benceno en los fondos (xB) es {xB}')
print(f'La fracción molar de Benceno en el destilado (xD) es {xD}')
```

```python
plt.figure(figsize=(10, 10))
plt.plot(X, Y, label='CDE')
plt.plot(aux_X, aux_Y, 'r', label='Línea aux 45°')
plt.vlines(x = zF, ymin = 0, ymax = zF, colors = 'purple', label='zF')
plt.vlines(x = xB, ymin = 0, ymax = xB, colors = 'purple', linestyles='dotted', label='xB')
plt.vlines(x = xD, ymin = 0, ymax = xD, colors = 'purple', linestyles='dashdot', label='xD')
plt.title('CDE')
plt.xlabel('X (Fracción molar de Benceno en el líquido')
plt.ylabel('Y (Fracción molar de Benceno en el gas')
plt.grid(True)
plt.legend()
```

# Paso 4

El enunciado del ejercicio nos da el valor de la relación equimolar de vapor y líquido: **0.5**. Con esto, se sabe que el ángulo de la línea q será de 45° respecto a la proyección de la línea zF (purple central). Se procede a trazar una línea que atraviese la CDE para posteriormente encontrar el punto de intersección.

```python
prueba_qX = (0.4, 0.6)  # Puntos X para línea q (proyección)
prueba_qY = (0.8, 0.6)  # Puntos Y para línea q (proyección)

plt.figure(figsize=(10, 10))
plt.plot(X, Y, label='CDE')
plt.plot(aux_X, aux_Y, 'r', label='Línea aux 45°')
plt.vlines(x = zF, ymin = 0, ymax = zF, colors = 'purple', label='zF')
plt.vlines(x = xB, ymin = 0, ymax = xB, colors = 'purple', linestyles='dotted', label='xB')
plt.vlines(x = xD, ymin = 0, ymax = xD, colors = 'purple', linestyles='dashdot', label='xD')
plt.plot(prueba_qX, prueba_qY, label='Línea q (Proyección)')
plt.title('CDE')
plt.xlabel('X (Fracción molar de Benceno en el líquido')
plt.ylabel('Y (Fracción molar de Benceno en el gas')
plt.grid(True)
plt.legend()
```

Con ayuda del graficador de matplotlib, se encuentra que el punto de la CDE donde intersecta con la línea q, tiene coordenadas **(0.487, 0.713)**. A continuación, se grafica correctamente la línea q.

```python
qX = [0.487, 0.6]   # Puntos X para línea q
qY = [0.713, 0.6]   # Puntos Y para línea q

plt.figure(figsize=(10, 10))
plt.plot(X, Y, label='CDE')
plt.plot(aux_X, aux_Y, 'r', label='Línea aux 45°')
plt.vlines(x = zF, ymin = 0, ymax = zF, colors = 'purple', label='zF')
plt.vlines(x = xB, ymin = 0, ymax = xB, colors = 'purple', linestyles='dotted', label='xB')
plt.vlines(x = xD, ymin = 0, ymax = xD, colors = 'purple', linestyles='dashdot', label='xD')
plt.plot(qX, qY, label='Línea q')
plt.title('CDE')
plt.xlabel('X (Fracción molar de Benceno en el líquido')
plt.ylabel('Y (Fracción molar de Benceno en el gas')
plt.grid(True)
plt.legend()
```

# Paso 5

```python
pinch_point = (qX[0], qY[0])                                # Coordenadas del Pinch Point
lin_op_sup = [(pinch_point[0], xD), (pinch_point[1], xD)]   # Puntos X,Y para línea de operación superior
lin_op_inf = [(xB, pinch_point[0]), (xB, pinch_point[1])]   # Puntos X,Y para línea de operación inferior

plt.figure(figsize=(10, 10))
plt.plot(X, Y, label='CDE')
plt.plot(aux_X, aux_Y, 'r', label='Línea aux 45°')
plt.vlines(x = zF, ymin = 0, ymax = zF, colors = 'purple', label='zF')
plt.vlines(x = xB, ymin = 0, ymax = xB, colors = 'purple', linestyles='dotted', label='xB')
plt.vlines(x = xD, ymin = 0, ymax = xD, colors = 'purple', linestyles='dashdot', label='xD')
plt.plot(qX, qY, label='Línea q')
plt.scatter(pinch_point[0], pinch_point[1], color='black', label='Pinch Point')
plt.plot(lin_op_sup[0], lin_op_sup[1], color='green')
plt.plot(lin_op_inf[0], lin_op_inf[1], color='green', label='Líneas de operación')
plt.title('CDE')
plt.xlabel('X (Fracción molar de Benceno en el líquido')
plt.ylabel('Y (Fracción molar de Benceno en el gas')
plt.grid(True)
plt.legend()
```

# Paso 6

Se procede a hallar la pendiente de la línea de operación superior graficada.

```python
m = (xD - pinch_point[1]) / (xD - pinch_point[0])   # Pendiente de la LdeOp sup
print(f'La pendiente de la línea de operación superior graficada es {round(m,3)}')
```

Se realiza el procedimiento necesario para calcular la pendiente real de la línea de operación superior teniendo en cuenta la razón de reflujo del proceso.

Rcmin.jpg

```python
Rc_min = m / (1 - m)
print(f'La relación del reflujo mínimo de la torre de destilación es {round(Rc_min, 5)}')

razon_refl = 1.28
Rc = razon_refl * razon_refl
print(f'La relación real del reflujo de la operación es {round(Rc, 5)}')

m_LdeO_sup = Rc / (1 + Rc)
print(f'La pendiente real de la línea de operación superior es {round(m_LdeO_sup, 5)}')
```

Se halla el punto intercepto de la línea de operación superior real a partir de la ecuación de una línea recta.

```python
m_real = m_LdeO_sup     # Pendiente real de la línea de operación superior real
b = xD - (xD*m_real)    # Punto intercepto de la línea de operación superior real
print(f'El punto intercepto de la línea de operación superior real es {round(b, 5)}')
```

A continuación, se grafica la proyección de la línea de operación superior real del proceso. Se sigue el mismo procedimiento al realizado anteriormente con la línea q, para dar con la gráfica de la LdeOp sup real.

```python
prueba_op_sup_real_X = np.arange(0, X[-1]+0.1, 0.1)    # Puntos X para gráfica de LdeOp sup real
prueba_op_sup_real_Y = (m_real * prueba_op_sup_real_X) + b    # Puntos Y para gráfica de LdeOp sup real

plt.figure(figsize=(10, 10))
plt.plot(X, Y, label='CDE')
plt.plot(aux_X, aux_Y, 'r', label='Línea aux 45°')
plt.vlines(x = zF, ymin = 0, ymax = zF, colors = 'purple', label='zF')
plt.vlines(x = xB, ymin = 0, ymax = xB, colors = 'purple', linestyles='dotted', label='xB')
plt.vlines(x = xD, ymin = 0, ymax = xD, colors = 'purple', linestyles='dashdot', label='xD')
plt.plot(qX, qY, label='Línea q')
plt.scatter(pinch_point[0], pinch_point[1], color='black', label='Pinch Point')
plt.plot(lin_op_sup[0], lin_op_sup[1], color='green')
plt.plot(lin_op_inf[0], lin_op_inf[1], color='green', label='Líneas de operación')
plt.plot(prueba_op_sup_real_X, prueba_op_sup_real_Y, color='black', label='LdeOp superior real (Proyección)')
plt.title('CDE')
plt.xlabel('X (Fracción molar de Benceno en el líquido')
plt.ylabel('Y (Fracción molar de Benceno en el gas')
plt.grid(True)
plt.legend()
```

El punto de intersección de la línea q con la LdeOp sup real es **(0.516, 0.684)**

```python
interseccion = (0.516, 0.684)
```

```python
op_sup_real_X = np.arange(interseccion[0], xD, 0.01)    # Puntos X para LdeOp sup real
op_sup_real_Y = (m_real * op_sup_real_X) + b            # Puntos Y para LdeOp sup real

plt.figure(figsize=(10, 10))
plt.plot(X, Y, label='CDE')
plt.plot(aux_X, aux_Y, 'r', label='Línea aux 45°')
plt.vlines(x = zF, ymin = 0, ymax = zF, colors = 'purple', label='zF')
plt.vlines(x = xB, ymin = 0, ymax = xB, colors = 'purple', linestyles='dotted', label='xB')
plt.vlines(x = xD, ymin = 0, ymax = xD, colors = 'purple', linestyles='dashdot', label='xD')
plt.plot(qX, qY, label='Línea q')
plt.scatter(pinch_point[0], pinch_point[1], color='black', label='Pinch Point')
plt.plot(lin_op_inf[0], lin_op_inf[1], color='green', label='Línea de operación inferior (no real)')
plt.plot(op_sup_real_X, op_sup_real_Y, color='black', label='LdeOp superior real')
plt.title('CDE')
plt.xlabel('X (Fracción molar de Benceno en el líquido')
plt.ylabel('Y (Fracción molar de Benceno en el gas')
plt.grid(True)
plt.legend()
```

# Paso 7

```python
op_inf_real_X = (xB, interseccion[0])   # Puntos X para LdeOp inf real
op_inf_real_Y = (xB, interseccion[1])   # Puntos Y para LdeOp inf real

plt.figure(figsize=(10, 10))
plt.plot(X, Y, label='CDE')
plt.plot(aux_X, aux_Y, 'r', label='Línea aux 45°')
plt.vlines(x = zF, ymin = 0, ymax = zF, colors = 'purple', label='zF')
plt.vlines(x = xB, ymin = 0, ymax = xB, colors = 'purple', linestyles='dotted', label='xB')
plt.vlines(x = xD, ymin = 0, ymax = xD, colors = 'purple', linestyles='dashdot', label='xD')
plt.plot(qX, qY, label='Línea q')
plt.scatter(pinch_point[0], pinch_point[1], color='black', label='Pinch Point')
plt.plot(op_inf_real_X, op_inf_real_Y, color='black')
plt.plot(op_sup_real_X, op_sup_real_Y, color='black', label='Línea de operación real')
plt.title('CDE')
plt.xlabel('X (Fracción molar de Benceno en el líquido')
plt.ylabel('Y (Fracción molar de Benceno en el gas')
plt.grid(True)
plt.legend()
```
